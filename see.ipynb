```python
# data[:, 2] are performance_all
# data[:, 3] are performance_selected
# data[:, 4] are performance_random


performance_all = data[:, 2]
performance_selected = data[:, 3]
performance_random = data[:, 4]

s_d = torch.abs(performance_selected - performance_all)
r_d = torch.abs(performance_random - performance_all)


print(torch.mean(s_d), torch.mean(r_d))
print(torch.var(s_d), torch.var(r_d))
```
